```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load(path):
    df = np.loadtxt(path)
    df = pd.DataFrame(df, columns=['brick_id', 'SX', 'SY', 'SZ', 'TX', 'TY']) 
    return df
```
